```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def add_support_resistance_levels(df, column='Close', window_size=20):
    # Rolling maximum for the resistance level
    df['Resistance'] = df[column].rolling(window=window_size, min_periods=1).max().shift(1)
    
    # Rolling minimum for the support level
    df['Support'] = df[column].rolling(window=window_size, min_periods=1).min().shift(1)
    
    return df

# Assuming 'data' is your DataFrame and 'Close' is the column with closing prices
data = pd.DataFrame({
    'Close': [100, 101, 102, 99, 98, 103, 105, 104, 102, 101, 100, 99, 98, 97, 96, 95, 94, 93, 92, 91, 90]
})

# Use the function to add support and resistance columns to the data
enriched_data = add_support_resistance_levels(data, 'Close', window_size=5)
plt.plot(data['Support'])
plt.plot(data['Resistance'])
plt.plot(data['Close'])

# print(enriched_data)
```
